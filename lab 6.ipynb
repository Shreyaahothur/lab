```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from scipy.spatial.distance import minkowski
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor

# Load the dataset
df = pd.read_csv("C:\\Users\\Shreyaa Hothur\\OneDrive\\Desktop\\5th Sem\\ML\\Project\\extracted_features_dense_15.csv")

# Extract the columns to create a scatter plot for
d1 = df['30']
d2 = df['40']

# Create a scatter plot
plt.figure(figsize=(8, 6))  # Set the figure size
plt.scatter(d1, d2, c='g', marker='o', label='Data Points')  # Customize scatter plot
plt.xlabel("X-axis Label")
plt.ylabel("Y-axis Label")
plt.title("Scatter Plot")
plt.grid(True)  # grid lines
plt.legend()  # explaination 

# Show the plot
plt.show()
```

```python
X = np.array(df['43']).reshape(-1, 1) # The independent variable (X)  in 2D
Y = np.array(df['0']) # The dependent variable (Y)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values based on the model
Y_pred = model.predict(X)

# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(Y, Y_pred)

# Print the MSE
print("Mean Squared Error:", mse)
```

```python

# Define feature data (X) and class labels (Y)
X = df.drop(columns=['class_name'])
Y = df['class_name']

# Replace class labels based on the specified condition
Y = [0 if i in range(3333, 3346) else 1 for i in Y]

# Split your data into a training set and a test set
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Create a logistic regression classifier
clf = LogisticRegression(max_iter=100000)
clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = clf.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python

# Create a Decision Tree Regressor
dt_regressor = DecisionTreeRegressor()
dt_regressor.fit(X, Y)

# Create a k-NN Regressor
knn_regressor = KNeighborsRegressor(n_neighbors=5)
knn_regressor.fit(X, Y)

# Make predictions with both models
dt_predictions = dt_regressor.predict(X)
knn_predictions = knn_regressor.predict(X)

# Calculate Mean Absolute Error for both models
mae_dt = mean_absolute_error(Y, dt_predictions)
mae_knn = mean_absolute_error(Y, knn_predictions)

# Print MAE for each model
print(f"Decision Tree MAE: {mae_dt}")
print(f"k-NN MAE: {mae_knn}")
```
